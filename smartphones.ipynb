```python
from ortools.linear_solver import pywraplp
from ortools.linear_solver.pywraplp import Solver, Variable, Objective
```

```python
#declara o solver: se PLI, usa-se SCIP; se PL, usa-se GLOP.
solver: Solver = pywraplp.Solver.CreateSolver('SCIP')

#declara um número muito grande para ser usado no modelo
infinity = solver.infinity()
```

```python
#declara as variáveis do modelo: IntVar para inteira; NumVar para fracionárias; BoolVar para binárias.
x: Variable = solver.IntVar(2,infinity,"unidade compradas do modelo a")
y: Variable = solver.IntVar(2,infinity,"unidade compradas do modelo b")
z: Variable = solver.IntVar(2,infinity,"unidade compradas do modelo c")
```

```python
#declara a função objetivo
objetivo: Objective = solver.Objective()
objetivo.SetCoefficient(x, 500)
objetivo.SetCoefficient(y, 700)
objetivo.SetCoefficient(z, 600)
objetivo.SetMaximization()
```

```python
#declara restrições
restricao_1 = solver.Add(2000 * x + 3000 * y + 2500 * z <= 25000)
restricao_2 = solver.Add(z - x >= 1)
restricao_3 = solver.Add(z - y >= 1) 
```

```python
status = solver.Solve()

if status == pywraplp.Solver.OPTIMAL:
    print("Solução Ótima Encontrada!")
    print(f"Unidades compradas do modelo A: {x.solution_value()}")
    print(f"Unidades compradas do modelo B: {y.solution_value()}")
    print(f"Unidades compradas do modelo C: {z.solution_value()}")
    print(f"Lucro máximo: {objetivo.Value()}")
elif status == pywraplp.Solver.INFEASIBLE:
    print("O modelo não possui solução viável.")
elif status == pywraplp.Solver.UNBOUNDED:
    print("O modelo é ilimitado.")
else:
    print("Otimização não foi concluída corretamente.")
```

```python
#exibe o modelo
print(solver.ExportModelAsLpFormat(False))
```
